```scala
import org.apache.log4j.{Level, Logger}
Logger.getLogger("org").setLevel(Level.OFF)

import $ivy.`org.apache.spark::spark-sql:3.1.1`
import org.apache.spark.sql._

val spark = {
  NotebookSparkSession.builder()
    .master("local[*]")
    .getOrCreate()
}

def sc = spark.sparkContext

import $ivy.`org.apache.spark::spark-mllib:3.1.1`
import org.apache.spark.mllib.linalg.{Matrix, Matrices}

Matrices.dense(3, 2, Array(1, 3, 5, 2, 4, 6))
```

```scala
val rdd = sc.parallelize(List(("a", 1), ("b", 1), ("a", 2), ("b", 2), ("a", 3)))
val groupedRDD = rdd.groupByKey()
val avgRDD = groupedRDD.mapValues(values => values.sum / values.size.toDouble)
avgRDD.collect().foreach(println)
```
